# 量化交易
- 定义： 是指以先进的数学模型替代人为的主观判断，利用计算机技术从庞大的历史数据中海选出能带来超额收益的多种“大概率”事件以制定策略。
- 优势
    - 降低时长波动给投资者情绪带来的影响
    - 避免在时长季度狂热或者悲观的情况下做出非理性的投资决策
    - 通过交易策略抽离决策前沿
- 特点
    - 普通交易者根据自身经验偏好进行投资
    - 量化交易通过将数据（行情历史、基本面信息及新闻咨询等）输入模型，利用计算机及统计学技术方法分析数据，产生交易信号进行交易决策
- 注意事项
    - 不要过度拟合交易系统
    - 避免实现太过复杂的交易策略
    - 避免逆势操作
    - 避免重仓交易
    - 避免对胜率的盲目追求
    - 确保交易策略的执行
    - 要有独立交易及对结果负责的信念
    - 从历史验证交易策略是否可行
    - 寻找交易策略的最优参数
    - 减少无意义的工作和干扰
    

### 量化模型
- 包含选股模块、择时模块、风险控制模块
- 决策： 传统投资使用定性分析，量化交易使用定量分析
    - 定量分析：通过统计调查法或实验法， 像自然科学那样建立研究假设，收集精确的数据资料，然后进行统计分析和检验的研究过程。定量分析作为一种古已有之但是没有被准确定位的思维方式， 其优势相对于定性分析的确很明显， 它把事物定义在了人类能理解的范围，由定量而定性。
    - 定性分析：是在反实证主义理论影响下形成的一种社会科学研究方法。它趋向于运用访问、观察和文献法收集资料， 并依据主观的理解和定性分析进行研究的过程。定性研究强调人类行为是一种有意义的行动，人们对社会现实的建构是在主体以及参与互动的他人对社会客体赋予意义的基础上共同完成的，日常生活中具有公识的规则或知识也由此产生。对这些规则或知识及其产生过程， 很难用定量的方式进行研究。定性分析就是对研究对象进行“质”的方面的分析，具体地说是运用归纳和演绎、分析与综合以及抽象与概括等方法， 对获得的各种材料进行思维加工，从而能去粗取精、去伪存真、由此及彼、由表及里，达到认识事物本质、揭示内在规律，由定性而定量。

### 量化交易的正确认识
- 不因循守旧，认为量化交易是邪门歪道
- 不要异想天开，认为量化交易有神奇的魔法
- 不要抱有不劳而获的幻想
- 投资： 需要一个相对比较大的概率优势目标，比较准确的预测结果，前期投入大量资源对目标进行研究分析，以获取丰厚投资汇报为目标，进行长期投资的行为。
- 量化交易很大程度上也算是投机。
- 不要盲目追求量化策略的复杂性
- 认清市场，认清自己，知己知彼，百战不殆
